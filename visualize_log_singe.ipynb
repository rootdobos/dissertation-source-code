```python
import pandas as pd

# Load the CSV
df = pd.read_csv("F:/logs/resnet_clam/20250305_220807_p5_b50/train_log.csv")

df.head()
```

```python
import matplotlib.pyplot as plt

# Plot loss and accuracy
fig, ax1 = plt.subplots()

ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss", color="tab:red")
ax1.plot(df["epoch"], df["loss"], label="Loss", color="tab:red")
ax1.tick_params(axis="y", labelcolor="tab:red")

ax2 = ax1.twinx()
ax2.set_ylabel("Accuracy", color="tab:blue")
ax2.plot(df["epoch"], df["accuracy"], label="Accuracy", color="tab:blue")
ax2.tick_params(axis="y", labelcolor="tab:blue")

fig.tight_layout()
plt.title("Loss vs Accuracy")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["epoch"], df["macro_f1"], label="Macro F1", marker="o")
plt.plot(df["epoch"], df["weighted_f1"], label="Weighted F1", marker="s")
plt.plot(df["epoch"], df["qwk"], label="Quadratic Weighted Kappa", marker="^")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Evaluation Metrics Over Epochs")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i in range(6):  # Assuming classes are class0 to class5
    plt.plot(df["epoch"], df[f"class{i}"], label=f"Class {i}")

plt.xlabel("Epoch")
plt.ylabel("Performance")
plt.title("Class-wise Performance")
plt.legend()
plt.grid()
plt.show()
```
