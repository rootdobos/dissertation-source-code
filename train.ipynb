```python
from deep_learning.loaders.feature_dataset import FeatureDataset
import pandas as pd
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
from deep_learning.train.train import train
from deep_learning.models.attention_core.clam import Clam

df=pd.read_csv("F:/data/small.csv")
train_df, test_df = train_test_split(df[:100], test_size=0.2, random_state=42)

model=Clam(
    feature_vector_length=2048,
    dropout=0.1,
    k_sample=8,
    n_classes=6,
    subtyping=False
)



train(model,train_df,"F:/extracted_features",10,1,"F:/logs/resnet_clam_small")
```
